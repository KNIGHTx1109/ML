```python
sf = pd.DataFrame()
```

```python
for i in range(1,500):
  dataa =  requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=dae59b7b274dc327d3d57da533d574cd&language=en-US&page={}'.format(i))
  temp_df = pd.DataFrame(dataa.json()['results'])[['title','release_date','popularity','vote_average','vote_count']]
  sf = pd.concat([sf, temp_df], ignore_index=True)
```
